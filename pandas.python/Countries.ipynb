```python
#how many Continent have south in their full namne
count = 0
def counting(txt):
    global count 
    if 'south' in txt.lower():
        count+=1
        return txt
df['Continent'] = df['Continent'].apply(counting)
print(count)
```

```python
# Filter only Asian countries
africa_df = df[df['Continent'] == 'Africa']

# Get the row(s) with the maximum population in Asia
largest_pop_africa = africa_df[africa_df['Population_millions'] == africa_df['Population_millions'].max()]['Country']

print(largest_pop_africa)
```

```python

# Check continent labels
print(df['Continent'].unique())

# Strip whitespace
df['Continent'] = df['Continent'].str.strip()

# Ensure numeric population
df['Population_millions'] = pd.to_numeric(df['Population_millions'], errors='coerce')

# Filter Asia again
asia_df = df[df['Continent'] == 'Asia']
print(asia_df.shape)   # should not be (0, x)

# Get largest population in Asia
largest_pop_asia = asia_df[asia_df['Population_millions'] == asia_df['Population_millions'].max()]
print(largest_pop_asia[['Country','Population_millions']])
```
