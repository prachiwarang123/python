```python
import pandas as pd
import numpy as np

# Number of rows (make it big!)
num_rows = 1000

# Example countries
countries = ["India", "USA", "Brazil", "Germany", "Nigeria", "Japan", "China", "UK", "France", "Canada"]

continents = {
    "India": "Asia", "USA": "North America", "Brazil": "South America",
    "Germany": "Europe", "Nigeria": "Africa", "Japan": "Asia",
    "China": "Asia", "UK": "Europe", "France": "Europe", "Canada": "North America"
}

capitals = {
    "India": "New Delhi", "USA": "Washington", "Brazil": "Bras√≠lia",
    "Germany": "Berlin", "Nigeria": "Abuja", "Japan": "Tokyo",
    "China": "Beijing", "UK": "London", "France": "Paris", "Canada": "Ottawa"
}

currencies = {
    "India": "INR", "USA": "USD", "Brazil": "BRL",
    "Germany": "EUR", "Nigeria": "NGN", "Japan": "JPY",
    "China": "CNY", "UK": "GBP", "France": "EUR", "Canada": "CAD"
}

languages = {
    "India": "Hindi", "USA": "English", "Brazil": "Portuguese",
    "Germany": "German", "Nigeria": "English", "Japan": "Japanese",
    "China": "Mandarin", "UK": "English", "France": "French", "Canada": "English"
}

areas = {
    "India": 3287263, "USA": 9833520, "Brazil": 8515767,
    "Germany": 357022, "Nigeria": 923769, "Japan": 377975,
    "China": 9596961, "UK": 243610, "France": 551695, "Canada": 9984670
}

# Generate dataset
data = []
for i in range(num_rows):
    country = np.random.choice(countries)
    year = np.random.randint(2000, 2024)
    population = np.random.randint(10, 1500)  # millions
    gdp = np.random.randint(100, 25000)       # billion USD
    timestamp = pd.Timestamp("2000-01-01") + pd.Timedelta(days=i)
    
    data.append([
        i+1,
        country,
        continents[country],
        population,
        gdp,
        year,
        capitals[country],
        currencies[country],
        languages[country],
        areas[country],
        timestamp.strftime("%Y-%m-%d %H:%M:%S")
    ])

# Build DataFrame
df = pd.DataFrame(data, columns=[
    "Rank", "Country", "Continent", "Population_millions", "GDP_billion_USD",
    "Year", "Capital", "Currency", "Language", "Area_sq_km", "Timestamp"
])

# Save to CSV
df.to_csv("large_country_data.csv", index=False)

print(df.head())
print(df.shape)
```
