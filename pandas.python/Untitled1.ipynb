```python
import pandas as pd
import numpy as np

# Create 50 ranks
ranks = list(range(1, 51))

# Generate timestamps (daily episodes starting from 2023-01-01 at 10:00 AM)
timestamps = pd.date_range("2023-01-01 10:00:00", periods=50, freq="D")

# Titles with episode numbers and timestamps inside
titles = [f"Taarak Mehta Ka Ooltah Chashmah (epi {i:02d}) {timestamps[i-1].strftime('%Y-%m-%d %H:%M:%S')}" for i in ranks]

# Random scores between 70 and 100
scores = np.random.randint(70, 101, size=50)

# Build DataFrame
df = pd.DataFrame({
    "Rank": ranks,
    "Title": titles,
    "Score": scores
})

# Save to CSV
df.to_csv("tmkoc_episodes_with_timestamp_in_title.csv", index=False)

print(df.head())  # preview first 5 rows
```
