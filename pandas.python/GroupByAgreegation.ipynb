```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Categories' : [ 'A', 'B','A', 'B','A', 'B','A', 'B'], # 8 values alternating
    'Stores': ['S1', 'S2', 'S1', 'S2','S1', 'S2','S1', 'S2'],    # 8 values alternating
    'Sales' : [100, 150, 120, 180, 130, 160, 140, 170],
    'Quantity' : [10, 15, 12, 18, 13, 16, 14, 17],
    'Dates' : pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Group by category and calculate the sum of sales
cat = df.groupby('Categories')
cat
for i, v in cat:
    print(i)
    print(v)
```

```python
cat = df.groupby('Categories')['Sales'].sum()
cat
```

```python
cat = df.groupby('Stores')['Sales'].sum()
cat
```

```python
#group by multiple columns
#group by categories and stores
cat = df.groupby(['Categories', 'Stores'])['Sales'].sum()
cat
```
