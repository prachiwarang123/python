```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == '(':
            check = True
            continue
        if i == ')':
            break
        if check:
            data += i
    return data
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
#for timestamp 
def extraction_time(txt):
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':  # find closing bracket
            # slice next 18 characters after ')'
            data = txt[i+1:i+19]
            break
    return data
```

```python
df['Title'].apply(extraction_time)
```

```python
df['Timestap'] = df['Title'].apply(extraction_time)
```

```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("tmkoc_episodes.csv")

# Step 1: Extract timestamp part (everything after ')')
df["Timestamp"] = df["Title"].str.split(")").str[1].str.strip()

# Step 2: Convert to datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"], format="%Y-%m-%d %H:%M:%S")

# Step 3: Extract month number and month name
df["Month_Number"] = df["Timestamp"].dt.month
df["Month_Name"] = df["Timestamp"].dt.month_name()

# Show the extracted months
print(df[["Title", "Month_Number", "Month_Name"]])
```
