```python
import numpy as np
import pandas as pd
```

```python
#Merging two data frames
```

```python
employees = pd.DataFrame({
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["John", "Prachi", "Aku", "Riya", "Rina"],
    "Department": ["IT", "Finance", "HR", "Sales", "Designer"]
})

salaries = pd.DataFrame({
    "EmployeeID": [101, 102, 103, 106, 107],
    "Salary": [65000, 90000, 70000, 60000, 50000],
    "Bonus": [5000, 12000, 7000, 5000, 3000]   # Bonus added for each employee
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries, on='EmployeeID', how = 'inner')
```

```python
pd.merge(employees,salaries, on='EmployeeID', how = 'outer')
```

```python
pd.merge(employees,salaries, on='EmployeeID', how = 'left')
```

```python
pd.merge(employees,salaries, on='EmployeeID', how = 'right')
```

```python
#concatination of 2 dataframes
```

```python
df1 = pd.DataFrame({
    'A':  ['A0', 'A1', 'A2'],
    'B':  ['B0', 'B1', 'B2'],
    'C':  ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A':  ['A3', 'A4', 'A5'],
    'B':  ['B3', 'B4', 'B5'],
    'C':  ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

```python
#Joining 2 dataframes
```

```python
df1 = pd.DataFrame(
    {"Name": ["John", "Prachi", "Aku"]},
    index=[1, 2, 3])

df2 = pd.DataFrame(
    {"Score": [85, 90, 75]},
    index=[2, 3, 4]
)
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how = 'outer')
```

```python
df2.join(df1)
```
