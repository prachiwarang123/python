```python
xaxis = np.linspace(0,5,11)
```

```python
y1 = np.sin(xaxis)
y2 = np.cos(xaxis)
y3 = np.tan(xaxis / 5)  # scaled to avoid extreme values
y4 = np.exp(-xaxis)     # exponential decay

# Create figure and axes
fig = plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot each line with a different color
axis1.plot(xaxis, y1, color='red', label='sin(x)')
axis1.plot(xaxis, y2, color='green', label='cos(x)')
axis1.plot(xaxis, y3, color='blue', label='tan(x/5)')
axis1.plot(xaxis, y4, color='purple', label='exp(-x)')

# Add legend, grid, and labels
axis1.legend()
axis1.grid(True)
axis1.set_title("Multiple Colored Lines")
axis1.set_xlabel("x-axis")
axis1.set_ylabel("y-axis")

# Save as JPEG
fig.savefig("multi_colored_graph.jpeg", format='jpeg', dpi=300)

plt.show()
```
